```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def calc_epsilon(s1, s2):
    return 0.25* min(abs(s1 - s2), s1, s2, 1 - s1, 1 - s2)
```

```python
# Create arrays for shades s1 and s2
s1, s2 = np.linspace(0, 1, 100), np.linspace(0, 1, 100)

# Create meshgrid
S1, S2 = np.meshgrid(s1, s2)

# Apply function on meshgrid
epsilons = np.vectorize(calc_epsilon)(S1, S2)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(S1, S2, epsilons, cmap='viridis')

# Set the viewing angle
ax.view_init(elev=28, azim=265) 

# Labels and title
ax.set_xlabel('Value for $s_1$')
ax.set_ylabel('Value for $s_2$')
ax.set_zlabel('Epsilon')
ax.set_title('Visualisation of the maximal permitted perturbation for two shades')

# Show the plot
plt.show()
```
